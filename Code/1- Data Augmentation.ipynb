```python
#!/usr/bin/python3
import os
import sys
import argparse

#####################################################
def load_NE_lists():
    ne_dict = {}
    ne_dir = '/kaggle/input/userstory-ner/'
    print('asdfas')
    for filename in os.listdir(ne_dir):
        #print(filename)
        if filename.endswith(".txt") and filename != 'innn.txt':
            # print(filename)
            ne_file = os.path.join(ne_dir, filename)
            entities = open(ne_file).read().splitlines()
            # list.Adj.txt 
            entity_name, ext = filename.split('.')
            ne_dict[entity_name] = entities
            print('{} loaded. NE: {}'.format(filename, entity_name))
    return ne_dict

#load_NE_lists()


#####################################################    
def get_more_named_entities(word, ne_dict):
    
    for ne_name, ne_list in ne_dict.items():
        ne_list_lower = []
        for ne_list_word in ne_list:
            ne_list_lower.append(ne_list_word.lower())
            
        if word in ne_list_lower:
            #print("wwwww: ", word)
            ne_list_lower.remove(word) # remove the word which already exist in the sentence.
            #To avoid duplicating the sentence
            #print("llll222: ", ne_list2)
            return ne_list_lower
    return None
    
#####################################################
def generate_sentences(NEs, words, index):
    gen_list = list()
    for ne in NEs:
        sen = list(words)
        sen[index] = ne
        gen_list.append(' '.join(sen).replace(' ,', ','))
    return gen_list
#####################################################

            

#####################################################        

#####################################################                    
                    
                    
#if __name__ == '__main__':
    #args = parser.parse_args()
ne_dict = load_NE_lists()
#input = '/kaggle/input/ner-data/innn.txt' #args.infile.readlines()
input = open('/kaggle/input/userstory-ner/innn.txt', "r")
#print (input)
file_out = open("test22.txt", "w")
out = file_out #args.outfile

for line in input:
    #print(line)
    words = line.strip().replace(',', ' ,').split()
    for i, word in enumerate(words):
        word = word.lower()
        NEs = get_more_named_entities(word, ne_dict)
        if NEs:
            new_sentences = generate_sentences(NEs, words, i)
            #print(new_sentences)
            out.writelines('\n'.join(new_sentences))
            out.write('\n')
out.close()           
print('expansion is done!')
```
